```python
columns = list(yt.columns)
print("Unique Values\n")

for col in columns:
    value =yt[col].nunique()
    print(f"{col} = {value}")
```

```python
yt[["trending_year","trending_date", "trending_month"]] = yt["trending_date"].str.split(".", expand =True)
```

```python
yt.drop(['video_error_or_removed'],inplace=True,axis = 1)
```

```python
yt.duplicated().sum()
```

```python
columns = list(yt.columns)
print('Null Vlaues\n')
for col in columns:
    value = yt[col].isna().sum()
    print(f"{col} = {value}")
```

# Data Visualise 

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
t_2017 = yt[yt["trending_year"] == '17']
t_2017 = t_2017.groupby('channel_title').agg({"views": "sum"}).sort_values(by = 'views', ascending = False).reset_index().head(5)
a_17 = sns.barplot(data= t_2017, x = "channel_title", y = "views", palette='coolwarm')
plt.bar_label(a_17.containers[0])
plt.title("Top 5 channel in 2017")

plt.subplot(1,2,2)
t_2018=yt[yt["trending_year"] == '18']
t_2018 = t_2018.groupby("channel_title").agg({"views":"sum"}).sort_values(by= "views", ascending = False).reset_index().head(5)
a_18 = sns.barplot(data = t_2018, x = "channel_title", y = "views", palette = "coolwarm")
plt.bar_label(a_18.containers[0])
plt.title("Top 5 channel  in2018")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(20,7))

plt.subplot(1,2,1)
data_2017 = yt[yt["trending_year"]== "17"]
data_2017 = data_2017.groupby("channel_title").agg({"views":"sum"}).sort_values(by = "views",ascending= True).reset_index().head(5)
t_2017 = sns.barplot(data = data_2017, x= "channel_title", y= "views",palette= "rocket")
plt.bar_label(t_2017.containers[0])
plt.title("Top 5 least viewed channel in 2017")

plt.subplot(1,2,2)
data_2018 = yt[yt["trending_year"]== "18"]
data_2018 = data_2018.groupby("channel_title").agg({"views":"sum"}).sort_values(by = "views",ascending= True).reset_index().head(5)
t_2018 = sns.barplot(data = data_2018, x= "channel_title", y= "views",palette= "rocket")
plt.bar_label(t_2018.containers[0])
plt.title("Top 5 least viewed channel in 2018")
```

# most video viewed in 2017 and 2018

```python
plt.figure(figsize=(19,5))

plt.subplot(1,2,1)
data = yt.groupby('title').agg({"likes":"sum"}).sort_values(by = "likes", ascending = False).reset_index().head(5)
a = sns.barplot(data = data, x ="title", y= "likes", palette = "hot")
plt.xticks(rotation=90)
plt.bar_label(a.containers[0])
plt.title("Top 5 likes videos")

plt.subplot(1,2,2)
data = yt.groupby("title").agg({"dislikes":"sum"}).sort_values(by="dislikes", ascending = False).reset_index().head(5)
a = sns.barplot(data = data, x = "title", y = "dislikes", palette = "coolwarm")
plt.xticks(rotation=90)
plt.bar_label(a.containers[0])
plt.title("Top 5 dislikes videos")

plt.show()

```

```python
data = yt.groupby(['trending_month', "trending_year"]).agg({"views":"sum","likes":"sum","dislikes":"sum", "comment_count":"sum"}).sort_values(by="trending_year").reset_index()

plt.figure(figsize=(20,5))
plt.subplot(2,2,1)
sns.lineplot(data = data, x ="trending_month", y = "likes", label= "likes")

plt.subplot(2,2,2)
```
